```python
import numpy
import pandas
import tensorflow
import keras
import sys
```

```python
# 使用當前資料預測後1小時降雨量 因此將資料向後取1位
Train_Data = I_am_Taipei_Weather_Data_TrainingFeature.values[0:-3]
Train_Label = I_am_Taipei_Weather_Data_TrainingLabel.values[1:-2]
print(Train_Data)
print(Train_Label)

Train_Data = Train_Data.astype(float)
print(Train_Data)
```

```python
# Basic Practice

#使用keras 模組創建instance 物件實例
Im_TmeSequence_Model = keras.models.Sequential()


#將產生的類神經網路物件 開始架構結構


#輸入層 input layer
Im_TmeSequence_Model.add(keras.layers.recurrent.LSTM(256, input_shape=(None, 4), return_sequences=True))
Im_TmeSequence_Model.add(keras.layers.core.Dropout(0.1))

Im_TmeSequence_Model.add(keras.layers.recurrent.LSTM(512, return_sequences=True))
Im_TmeSequence_Model.add(keras.layers.core.Dropout(0.2))

Im_TmeSequence_Model.add(keras.layers.recurrent.SimpleRNN(256, return_sequences=True))
Im_TmeSequence_Model.add(keras.layers.core.Dropout(0.2))

Im_TmeSequence_Model.add(keras.layers.core.Dense(units = 512, activation = "relu"))
Im_TmeSequence_Model.add(keras.layers.core.Dropout(0.2))

Im_TmeSequence_Model.add(keras.layers.recurrent.LSTM(256, return_sequences=True))
Im_TmeSequence_Model.add(keras.layers.core.Dropout(0.2))

Im_TmeSequence_Model.add(keras.layers.recurrent.LSTM(128, return_sequences=True))
Im_TmeSequence_Model.add(keras.layers.core.Dropout(0.1))

# 最後一層 = 輸出層 output layer
Im_TmeSequence_Model.add(keras.layers.core.Dense(units = 1, activation = "linear"))
# 定義訓練優化方法
Im_AdamAMSgrad_Optimizer = keras.optimizers.Adam(amsgrad=True)
# 產生模型
Im_TmeSequence_Model.compile(loss = "mse", optimizer = Im_AdamAMSgrad_Optimizer )
```

```python
TmeSequence_Train_Data = numpy.reshape(Train_Data,(1,1511,4))
TmeSequence_Train_Label = numpy.reshape(Train_Label,(1,1511,1))

# 使用model訓練
history = Im_TmeSequence_Model.fit(TmeSequence_Train_Data, TmeSequence_Train_Label, epochs=50, batch_size=1)
print(history)

#
loss_and_metrics = Im_TmeSequence_Model.evaluate(TmeSequence_Train_Data, TmeSequence_Train_Label)
print(loss_and_metrics)

# 使用後半部份驗證資料 20天 24 小時降雨量
Test_Data = I_am_Taipei_Weather_Data_TrainingFeature.values[-482:-2]
TmeSequence_Test_Data = numpy.reshape(Test_Data,(1,480, 4))


# 使用model預測 20天 24 小時降雨量
RainVolumn_Prediction = Im_TmeSequence_Model.predict(TmeSequence_Test_Data)
print("Predict 20 day rain : ")
print(RainVolumn_Prediction)
# 實際資料之 20天 24 小時降雨量
print("Real 20 day rain : ")
print(I_am_Taipei_Weather_Data_TrainingLabel.values[-482:-2])
```
