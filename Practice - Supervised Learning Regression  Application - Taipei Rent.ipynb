```python
import numpy
import pandas
import tensorflow
import keras
import sys
```

```python
# Basic Practice

#使用keras 模組創建instance 物件實例
Im_model = keras.models.Sequential()

#將產生的類神經網路物件 開始架構結構

#輸入層 input layer
Im_model.add(keras.layers.Dense(units=64, activation='relu', input_dim=5))

Im_model.add(keras.layers.Dense(units=128, activation='relu'))

Im_model.add(keras.layers.Dense(units=64, activation='relu'))

Im_model.add(keras.layers.Dense(units=128, activation='relu'))

#最後一層 = 輸出層 output layer
Im_model.add(keras.layers.Dense(units=1, activation='linear'))

Im_model.compile(loss='mse',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
Train_Data = I_am_Taipei_Renting_Data_TrainingData.values[1:]
Train_Label = I_am_Taipei_Renting_Data_TrainingLabel.values[1:]

Test_Data = I_am_Taipei_Renting_Data_BeitouDistric_TrainingData.values[:]
Test_Label = I_am_Taipei_Renting_Data_BeitouDistric_TrainingLabel.values[:]

# 使用model訓練
history = Im_model.fit(Train_Data, Train_Label, epochs=300, batch_size=1)
print(history)

# 
loss_and_metrics = Im_model.evaluate(Test_Data, Test_Label)
print(loss_and_metrics)

# 使用model預測
classes = Im_model.predict(Test_Data)
print(classes)
```
